```python
# Block 2
#This script came from class, it was repurposed on my end to calculate the vader sentiment and textblob sentiment for each comment in a youtube video
# Set your API key here
API_KEY = 'AIzaSyCvXI6pWC6v1Awkgh61ConDAMxyzHUKuB4'

nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('vader_lexicon')

sentimentAnalyser = SentimentIntensityAnalyzer()

def calculate_sentiment_vader(text):
    # Run VADER on the text
    scores = sentimentAnalyser.polarity_scores(text)
    # Extract the compound score
    compound_score = scores['compound']
    # Return compound score
    return compound_score

def calculate_sentiment_tb(text):
    # Run TextBlob sentiment analysis on the text
    blob = TextBlob(text)
    # Extract the polarity score (ranges from -1 to 1)
    polarity_score = blob.sentiment.polarity
    # Return the polarity score
    return polarity_score

def save_to_dataframe(comment_threads):
    """
    Convert the list of comment threads into a Pandas DataFrame.

    :param comment_threads: A list of comment threads in dictionary format.
    :return: A Pandas DataFrame containing comments and replies.
    """
    df = pd.DataFrame(comment_threads)
    return df

def get_wordnet_pos(treebank_tag):
    """
    Map Treebank POS tags to WordNet POS tags.

    :param treebank_tag: Treebank POS tag.
    :return: Corresponding WordNet POS tag.
    """
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun

def preprocess_text(text):
    """
    Preprocess text by removing punctuation, stopwords, and lemmatizing.

    :param text: Input text.
    :return: Preprocessed text.
    """
    # Make text lowercase
    text = text.lower()

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Tokenize the text
    words = word_tokenize(text)

    # Remove English stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Perform POS tagging
    pos_tags = nltk.pos_tag(words) # use this code or line below not both.
    #pos_tags = nltk.pos_tag(filtered_words)

    # Lemmatize the words with specified POS tags
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]

    return ' '.join(lemmatized_words)

def extract_comment_threads(video_id, max_results=100):
    """
    Extract comment threads, video title, comment dates, and reply dates from a YouTube video using the YouTube Data API.

    :param video_id: The ID of the YouTube video.
    :param max_results: The maximum number of comments to retrieve (default is 100).
    :return: A list of comment threads, where each thread includes video title, comment dates, reply dates, original comment, and replies.
    """
    youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

    # Get video details
    video_response = youtube.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    video_title = video_response['items'][0]['snippet']['title']

    # Get video comments
    comment_threads = []
    nextPageToken = None
    while len(comment_threads) < max_results:
        response = youtube.commentThreads().list(
            part='snippet,replies',
            videoId=video_id,
            maxResults=100,
            textFormat='plainText',
            pageToken=nextPageToken
        ).execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comment_text = comment['textDisplay']
            comment_date = comment['publishedAt']
            replies = []
            if 'replies' in item:
                for reply_item in item['replies']['comments']:
                    reply = reply_item['snippet']
                    reply_text = reply['textDisplay']
                    reply_date = reply['publishedAt']
                    replies.append({'text': reply_text, 'date': reply_date})

            # Preprocess comment and reply text
            comment_text = preprocess_text(comment_text)
            replies = [{'text': preprocess_text(reply['text']), 'date': reply['date']} for reply in replies]

            comment_threads.append({
                'video_title': video_title,
                'comment_date': comment_date,
                'comment': comment_text,
                'replies': replies
            })

        if 'nextPageToken' in response:
            nextPageToken = response['nextPageToken']
        else:
            break

    return comment_threads
```

```python
#Block 4
#This is the chat GPT output I got from asking it to loop the video ID's. This reduces the lines of code needed, because instead of doing 1,2,3...15. I just do it once with all the links in a list of strings.
#This process will be repeated once for each CEO rather than multiple times for each CEO.
#I also double checked the average sentiment values of my code vs the generated code and found that they resulted in the same value.

import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import googleapiclient.discovery

# Initialize sentiment analyzer
sentimentAnalyser = SentimentIntensityAnalyzer()

# Function to calculate VADER sentiment
def calculate_sentiment_vader(text):
    scores = sentimentAnalyser.polarity_scores(text)
    return scores['compound']

# Function to calculate TextBlob sentiment
def calculate_sentiment_tb(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Preprocessing text function (assuming defined earlier)
def preprocess_text(text):
    """
    Preprocess text by removing punctuation, stopwords, and lemmatizing.

    :param text: Input text.
    :return: Preprocessed text.
    """
    # Make text lowercase
    text = text.lower()

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # Tokenize the text
    words = word_tokenize(text)

    # Remove English stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Perform POS tagging
    pos_tags = nltk.pos_tag(words) # use this code or line below not both.
    #pos_tags = nltk.pos_tag(filtered_words) dont use stop words, removes words like not, which change the meaning of a sentence.

    # Lemmatize the words with specified POS tags
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]

    return ' '.join(lemmatized_words)

# Function to extract comment threads from a video
def extract_comment_threads(video_id, max_results=100):
    youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=API_KEY)

    # Get video details
    video_response = youtube.videos().list(part='snippet', id=video_id).execute()
    video_title = video_response['items'][0]['snippet']['title']

    # Get video comments
    comment_threads = []
    nextPageToken = None
    while len(comment_threads) < max_results:
        response = youtube.commentThreads().list(
            part='snippet,replies',
            videoId=video_id,
            maxResults=100,
            textFormat='plainText',
            pageToken=nextPageToken
        ).execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comment_text = comment['textDisplay']
            comment_date = comment['publishedAt']

            replies = []
            if 'replies' in item:
                for reply_item in item['replies']['comments']:
                    reply = reply_item['snippet']
                    reply_text = reply['textDisplay']
                    reply_date = reply['publishedAt']
                    replies.append({'text': reply_text, 'date': reply_date})

            # Preprocess comment and reply text
            comment_text = preprocess_text(comment_text)
            replies = [{'text': preprocess_text(reply['text']), 'date': reply['date']} for reply in replies]

            comment_threads.append({
                'video_title': video_title,
                'video_id': video_id,  # Add video ID to each entry
                'comment_date': comment_date,
                'comment': comment_text,
                'replies': replies
            })

        if 'nextPageToken' in response:
            nextPageToken = response['nextPageToken']
        else:
            break

    return comment_threads

# Function to process multiple video IDs
def process_videos(video_ids, max_results=100):
    all_dataframes = []
    results = []

    for video_id in video_ids:
        print(f"Processing video ID: {video_id}")
        comment_threads = extract_comment_threads(video_id, max_results)

        # Convert to DataFrame
        df = pd.DataFrame(comment_threads)

        # Add sentiment scores
        df['sentiment_score_vader'] = df['comment'].apply(calculate_sentiment_vader)
        df['sentiment_score_tb'] = df['comment'].apply(calculate_sentiment_tb)

        # Calculate average sentiments
        vader_avg = np.mean(df['sentiment_score_vader'][df['sentiment_score_vader'] != 0])
        tb_avg = np.mean(df['sentiment_score_tb'][df['sentiment_score_tb'] != 0])

        results.append({
            'video_id': video_id,
            'vader_avg_sentiment': vader_avg,
            'textblob_avg_sentiment': tb_avg
        })

        # Collect DataFrame
        all_dataframes.append(df)

    # Combine all DataFrames into one
    combined_df = pd.concat(all_dataframes, ignore_index=True)

    # Create a summary DataFrame for average sentiments
    summary_df = pd.DataFrame(results)

    return combined_df, summary_df

# Main script
if __name__ == "__main__":
    # List of video IDs
    Pat_Gelsinger = ['lLvDoSaU8O8', 'bBy2SwtLSTc', 'bCuAeu_YXB8', 'FSVxdROoadc','PSGjUSR0F_M', 'l9VwXN0V0g4', 'Sh5D9dC0X3U', 'oT2xNQdfxBo', '1lmXwSw7kqM', 'bWcN4a62i0Q', 'z23OWa0t4PM', 'AZPrENlhUT0', 'WXfjGWlz-Nc', 'isTX2hTQaA4', 'axPAuHTAo0Q']
    max_results = 10000  # Adjust as needed

    # Process videos
    comments_df, sentiment_summary_df = process_videos(Pat_Gelsinger, max_results)

    # Display results
    print("Combined Comments DataFrame:")
    print(comments_df.head())
    print("\nSentiment Summary DataFrame:")
    print(sentiment_summary_df)
```

```python
# Gather data for two-sample T-test
# Count number scores that are not zero. Zeroes represent comments that couldn't be analyzed or comments that may be irrelevant to the CEO.
Pat_Sample_Size =np.count_nonzero(Pat_All_scores)
Ginny_Sample_Size =np.count_nonzero(Ginny_All_scores)
Hock_Sample_Size = np.count_nonzero(Hock_All_scores)
Lisa_Sample_Size = np.count_nonzero(Lisa_All_scores)
Shantanu_Sample_Size = np.count_nonzero(Shantanu_All_scores)
Jayshree_Sample_Size = np.count_nonzero(Jayshree_All_scores)

#Get average of all non-zero scores
Pat_Mean = np.mean(Pat_All_scores[Pat_All_scores != 0])
Ginny_Mean = np.mean(Ginny_All_scores[Ginny_All_scores != 0])
Hock_Mean = np.mean(Hock_All_scores[Hock_All_scores != 0])
Lisa_Mean = np.mean(Lisa_All_scores[Lisa_All_scores != 0])
Shantanu_Mean = np.mean(Shantanu_All_scores[Shantanu_All_scores != 0])
Jayshree_Mean = np.mean(Jayshree_All_scores[Jayshree_All_scores != 0])

import statistics

# Get Variance of each data set
Pat_Variance=statistics.variance(Pat_All_scores)
Ginny_Variance=statistics.variance(Ginny_All_scores)
Hock_Variance=statistics.variance(Hock_All_scores)
Lisa_Variance=statistics.variance(Lisa_All_scores)
Shantanu_Variance=statistics.variance(Shantanu_All_scores)
Jayshree_Variance=statistics.variance(Jayshree_All_scores)

# Find T value between male and female counter parts
T_Pat_Ginny = (Pat_Mean - Ginny_Mean) / np.sqrt((Pat_Variance / Pat_Sample_Size) + (Ginny_Variance / Ginny_Sample_Size))
T_Hock_Lisa = (Hock_Mean - Lisa_Mean) / np.sqrt((Hock_Variance / Hock_Sample_Size) + (Lisa_Variance / Lisa_Sample_Size))
T_Shantanu_Jayshree = (Shantanu_Mean - Jayshree_Mean) / np.sqrt((Shantanu_Variance / Shantanu_Sample_Size) + (Jayshree_Variance / Jayshree_Sample_Size))

DOF_Pat_Ginny = Pat_Sample_Size + Ginny_Sample_Size - 2
DOF_Hock_Lisa = Hock_Sample_Size + Lisa_Sample_Size - 2
DOF_Shantanu_Jayshree = Shantanu_Sample_Size + Jayshree_Sample_Size - 2


```

```python
#Block 5
# Function to scrape an article from a URL, I created my own code from what was given in class and asked chat gpt to loop through a list of articles for each CEO.
def scrape_article(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        article = soup.get_text()
        return article
    else:
        print(f"Failed to retrieve the article. HTTP Status Code: {response.status_code}")
        return ""

# Function to clean a single sentence
def clean_sentence(sentence):
    # Tokenize the sentence
    tokens = word_tokenize(sentence)

    # Convert tokens to lowercase and filter alphanumeric tokens
    tokens_lower = [t.lower() for t in tokens if t.isalnum()]

    # Remove stopwords
    stops = set(stopwords.words('english'))
    tokens_stops = [t for t in tokens_lower if t not in stops]

    # Lemmatize tokens
    lemmatizer = WordNetLemmatizer()
    tokens_clean = [lemmatizer.lemmatize(t) for t in tokens_stops]

    # Join the cleaned tokens back into a sentence
    cleaned_sentence = " ".join(tokens_clean)
    return cleaned_sentence

# Function to analyze sentiment for multiple articles and return DataFrames
def analyze_sentiment_multiple_articles(ceo_name, links):
    # Initialize SentimentIntensityAnalyzer for VADER
    vader = Vader()

    # Collect data for DataFrames
    sentence_data = []
    article_data = []

    for url in links:
        print(f"Processing URL: {url}")

        # Scrape the article
        article_text = scrape_article(url)

        # Skip if the article could not be retrieved
        if not article_text:
            continue

        # Split the text into sentences
        sentences = sent_tokenize(article_text)

        # Ignore the first 8 and last 7 sentences
        if len(sentences) > 15:  # Ensure enough sentences remain after slicing
            sentences = sentences[8:-7]
        else:
            print(f"Skipping article due to insufficient sentences: {url}")
            continue

        # Sentiment scores for VADER and TextBlob
        sentiment_scores_vader = []
        sentiment_scores_textblob = []

        for i, sentence in enumerate(sentences):
            cleaned_sentence = clean_sentence(sentence)

            # VADER sentiment score
            score_vader = vader.polarity_scores(cleaned_sentence)['compound']
            if score_vader != 0:
                sentiment_scores_vader.append(score_vader)

            # TextBlob sentiment score
            score_textblob = TextBlob(cleaned_sentence).sentiment.polarity
            if score_textblob != 0:
                sentiment_scores_textblob.append(score_textblob)

            # Add sentence-level data
            sentence_data.append({
                "CEO Name": ceo_name,
                "Article URL": url,
                "Sentence Number": i + 1,
                "Sentence (Cleaned)": cleaned_sentence,
                "Sentence Score (VADER)": score_vader,
                "Sentence Score (TextBlob)": score_textblob
            })

        # Calculate average sentiment scores for the article
        avg_sentiment_vader = sum(sentiment_scores_vader) / len(sentiment_scores_vader) if sentiment_scores_vader else None
        avg_sentiment_textblob = sum(sentiment_scores_textblob) / len(sentiment_scores_textblob) if sentiment_scores_textblob else None

        # Add article-level data
        article_data.append({
            "Article URL": url,
            "Average Sentiment Score (VADER)": avg_sentiment_vader,
            "Average Sentiment Score (TextBlob)": avg_sentiment_textblob
        })

    # Convert sentence-level and article-level data to DataFrames
    sentence_df = pd.DataFrame(sentence_data)
    article_df = pd.DataFrame(article_data)
    return sentence_df, article_df

# Main script
if __name__ == "__main__":
    # Links for Pat Gelsinger
    Pat_Links = [
        "https://edition.cnn.com/2024/12/02/tech/intel-ceo-pat-gelsinger-resigns/index.html",
        "https://www.cnn.com/2024/12/03/business/intel-ceo-gelsinger-ten-million-dollars/index.html",
        "https://www.cnn.com/2024/12/04/tech/intel-brand-decline-gelsinger-retirement/index.html",
        "https://edition.cnn.com/2023/01/17/tech/intel-ceo-chip-manufacturing-davos/index.html",
        "https://www.cnn.com/2021/12/08/perspectives/intel-ceo-chip-shortage/index.html",
        "https://www.cnn.com/2021/01/21/tech/intel-new-ceo-pat-gelsinger-outlook/index.html",
        "https://www.cnn.com/2021/01/13/investing/intel-new-ceo-pat-gelsinger/index.html",
        "https://www.cnn.com/2023/08/17/business/intel-tower-semiconductor-deal-china-block-intl-hnk/index.html",
        "https://www.cnn.com/2024/03/20/politics/chips-act-intel-funding-biden/index.html"
    ]

    # Analyze sentiment for Pat Gelsinger's articles
    pat_sentence_df, pat_article_df = analyze_sentiment_multiple_articles("Pat Gelsinger", Pat_Links)

    # Display the DataFrames
    print("\nSentence-Level Sentiment Analysis DataFrame:")
    print(pat_sentence_df)

    print("\nArticle-Level Average Sentiment DataFrame:")
    print(pat_article_df)

    # Save the DataFrames to CSV files (optional)
    sentence_output_csv = "/content/pat_sentence_analysis.csv"
    article_output_csv = "/content/pat_article_analysis.csv"

    pat_sentence_df.to_csv(sentence_output_csv, index=False, encoding='utf-8')
    pat_article_df.to_csv(article_output_csv, index=False, encoding='utf-8')

    print(f"\nSentence-Level DataFrame saved to: {sentence_output_csv}")
    print(f"Article-Level DataFrame saved to: {article_output_csv}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# List of dataframes
dataframes = [
    pat_sentence_df,
    ginny_sentence_df,
    hock_sentence_df,
    lisa_sentence_df,
    shantanu_sentence_df,
    jayshree_sentence_df
]

# Dictionary to store results
top_tfidf_results = {}

# Function to filter top 3 non-numeric and non-excluded TF-IDF terms
def get_top_tfidf(df, tfidf_matrix, feature_names, exclude_terms):
    top_results = {}
    for i, url in enumerate(df['Article URL']):
        # Extract row of TF-IDF values
        tfidf_row = tfidf_matrix[i].toarray().flatten()

        # Create a DataFrame of terms and their TF-IDF values
        terms_df = pd.DataFrame({
            'Term': feature_names,
            'TF-IDF': tfidf_row
        })

        # Filter out numeric terms and excluded terms
        terms_df = terms_df[
            ~terms_df['Term'].str.contains(r'\d') &
            ~terms_df['Term'].isin(exclude_terms)
        ]

        # Get top 3 terms by TF-IDF
        top_terms = terms_df.nlargest(3, 'TF-IDF')
        top_results[url] = top_terms
    return top_results

# Process each dataframe
exclude_terms = ['cnn', 'playing', 'source', 'video', 'ad', 'feedback']
for df_name, df in zip(
    ['pat', 'ginni', 'hock', 'lisa', 'shantanu', 'jayshree'],
    dataframes
):
    # Group by Article URL and combine sentences into a single string
    grouped = df.groupby('Article URL')['Sentence (Cleaned)'].apply(' '.join).reset_index()

    # TF-IDF Vectorization
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(grouped['Sentence (Cleaned)'])

    # Get feature names
    feature_names = vectorizer.get_feature_names_out()

    # Get top 3 non-numeric, non-excluded TF-IDF terms for each article
    top_tfidf_results[df_name] = get_top_tfidf(grouped, tfidf_matrix, feature_names, exclude_terms)

# Display the top 3 terms for each article in each dataframe
for df_name, results in top_tfidf_results.items():
    print(f"Top 3 TF-IDF terms for {df_name} DataFrame:")
    for url, top_terms in results.items():
        print(f"Article URL: {url}")
        print(top_terms)
        print()

```
